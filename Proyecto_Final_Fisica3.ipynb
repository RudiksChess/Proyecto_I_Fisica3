```
import matplotlib.pyplot as plt
import numpy as np
```

```
def espectrometro_1_particula(center_x, center_y, radio1):

    r = np.linspace(-np.pi/2, np.pi/2, 500)
    x_1 = radio_1*np.cos(r)
    y_1 = radio_1*np.sin(r)-radio_1
    
    plt.gca().set_aspect('equal', adjustable='box')
    plt.plot(x_1, y_1, 'r')
    
    plt.show()

def espectrometro_2_particula(center_x, center_y, radio1,radio2):
    r = np.linspace(-np.pi/2, np.pi/2, 500)
    x_1 = radio_1*np.cos(r)
    y_1 = radio_1*np.sin(r) -radio_1

    x_2 = radio_2*np.cos(r)
    y_2 = radio_2*np.sin(r)-radio_2
    
    plt.gca().set_aspect('equal', adjustable='box')
    plt.plot(x_1, y_1, 'r')
    plt.plot(x_2, y_2, 'b')
    plt.show()

def espectrometro_3_particula(center_x, center_y, radio1,radio2,radio3):
    r = np.linspace(-np.pi/2, np.pi/2, 500)
    x_1 = radio_1*np.cos(r)
    y_1 = radio_1*np.sin(r)-radio_1
    x_2 = radio_2*np.cos(r)
    y_2 = radio_2*np.sin(r)-radio_2
    x_3 = radio_3*np.cos(r)
    y_3 = radio_3*np.sin(r)-radio_3

   
    plt.gca().set_aspect('equal', adjustable='box')
    plt.plot(x_1, y_1, 'r')
    plt.plot(x_2, y_2, 'b')
    plt.plot(x_3, y_3, 'g')

    plt.show()

```

```
def creador_de_particulas(nombre, proton, neutron): 
   masa =masa_proton*proton+masa_neutron*neutron
   datos_particulas.append([nombre,str(masa)])

def radio(m,E,B):
  r= (m*E)/(float(masa_elementaria)*(B**2))
  return r

def masa(particula):
  particula_analizada = datos_particulas[particula]
  masa_analizada = float(particula_analizada[1])
  return masa_analizada
```
